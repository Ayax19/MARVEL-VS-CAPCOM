```python
#  DESCARGAR EL TEXTO
from nltk.corpus import gutenberg
```

```python
raw=gutenberg.raw('shakespeare-macbeth.txt')
```

```python
data=nltk.word_tokenize(raw)
```

```python
data
```

```python
# OBTENER LA FRECUENCIA DE LAS PALABRAS
FRECUENCIA=FreqDist(data)
FRECUENCIA
```

```python
# DISTRIBUCION DE LAS PALABRAS
FRECUENCIA.plot(50,cumulative=True)
```

```python
# RELACIÓN ENTRE LAS PALABRAS MAS FRECUENTES Y EL TITULO DEL LIBRO
FRECUENCIA.plot(25,cumulative=False)
```

```python
# LA UNICA PALABRA NO IRRELEVANTE QUE APARECE ENTRE LOS 25 MAS FRECUENTES ES "Macb"
MACB=data.count("Macb")
TEXTO=len(data)
RELACION=MACB/TEXTO
RELACION
```
